### 4锔 Limpieza de datos

```python
#Se crear copias independientes para no modificar los originales

ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()
```

```python
print(ventas_clean.head(10))
print(clientes_clean.head(10))
print(marketing_clean.head(10))
```

```python
# Funcion que limpia un dataframe eliminando duplicados, caracteres no deseado y filas duplicadas

def limpiar_datos(df, nombre):

    print("\n" + f" {nombre.upper()} ".center(60, "=") + "\n")

    # --- Eliminar duplicados ---
    duplicados_antes = df.duplicated().sum()
    df = df.drop_duplicates().reset_index(drop=True)
    duplicados_despues = df.duplicated().sum()
    print(f"Filas duplicadas eliminadas: {duplicados_antes - duplicados_despues}")

    # --- Limpiar columnas de texto ---
    columnas_texto = df.select_dtypes(include=["object"]).columns
    if len(columnas_texto) > 0:
        print(f"Columnas de texto a limpiar: {list(columnas_texto)}")
        for col in columnas_texto:
            # Eliminar espacios al inicio y final
            df[col] = df[col].str.strip()

            # Reemplazar caracteres no deseados (mantiene letras, n煤meros y signos comunes)
            df[col] = df[col].str.replace(r"[^a-zA-Z0-9谩茅铆贸煤帽\s.,-]", "", regex=True)
        print("Limpieza aplicada con .str.strip() y .str.replace().")
    else:
        print("No se encontraron columnas de texto para limpiar.")

    # --- Mostrar resultados ---
    print(f"\n{'Filas totales:':<10} {len(df)}")
    print(f"{'Columnas totales:':<10} {len(df.columns)}")

    return df


# Aplicar la limpieza a los tres datasets
ventas_clean = limpiar_datos(ventas_clean, "ventas")
clientes_clean = limpiar_datos(clientes_clean, "clientes")
marketing_clean = limpiar_datos(marketing_clean, "marketing")
```

```python
# Normalizar fechas

# Si alguna columna contiene fechas (por ejemplo "fecha" o "fechanotificacion"), se convertie a formato datetime de pandas
# to_datetime intenta interpretar el formato y transforma valores inv谩lidos en NaT (Not a Time).

for df in [ventas_clean, clientes_clean, marketing_clean]:
    for col in df.columns:
        if "fecha" in col.lower():
            df[col] = pd.to_datetime(df[col], format="%d%m%Y", errors="coerce")
            df[col] = df[col].dt.strftime("%d-%m-%Y")
```

```python
# Normalizar valores num茅ricos

if "precio" in ventas_clean.columns:

    ventas_clean["precio"] = (
        ventas_clean["precio"]
        .astype(str)                        # Convierte todo a texto
        .str.replace("$", "", regex=False)  # Elimina el s铆mbolo $
        #   "$"  texto literal a reemplazar
        #   ""   nuevo valor (vac铆o)
        #   regex=False  interpreta "$" literalmente
        .str.replace(",", "", regex=False)  # Elimina comas de miles 1,000  1000
        .str.strip()                        # Quita espacios sobrantes
    )
    ventas_clean["precio"] = pd.to_numeric(ventas_clean["precio"], errors="coerce")
    # pd.to_numeric convierte texto a n煤mero (float o int)

ventas_clean["cantidad"] = (
    ventas_clean["cantidad"]
    .fillna(0)        # reemplaza NaN por 0
    .astype(int)      # convierte a entero
)
```

```python
# ============================================
#  REPORTE GLOBAL DE CALIDAD DE DATOS
# ============================================
# Esta funci贸n lee los tres DataFrames limpios (o los recibe en memoria)
# y muestra un resumen comparativo de nulos, duplicados y tipos de datos.

def reporte_calidad_global(dfs, nombres):
    """
    Crea un resumen de calidad de varios DataFrames.

    Par谩metros:
      dfs: lista de DataFrames (por ejemplo [ventas_clean, clientes_clean, marketing_clean])
      nombres: lista de nombres correspondientes (["VENTAS", "CLIENTES", "MARKETING"])
    """
    resumen = []
    #zip-->es una funci贸n incorporada de Python que une elementos de dos (o m谩s) iterables
    # por ejemplo, listas, tuplas o cualquier objeto iterable en pares ordenados.
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()                    # Total de valores nulos, no por columnas sino total, por eso el doble sum
        duplicados = df.duplicated(keep=False).sum()     # Total de filas duplicadas
        columnas = len(df.columns)                       # Cantidad de columnas
        filas = len(df)                                  # Cantidad de registros

        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)
    #display(reporte)
    return reporte
```

```python
print(reporte_calidad_global([ventas, clientes, marketing], ["VENTAS Original", "CLIENTES Original", "MARKETING Original"]))
print(reporte_calidad_global([ventas_clean, clientes_clean, marketing_clean],["VENTAS Copia   ", "CLIENTES Copia   ", "MARKETING Copia   "]))
```

### 5锔 Transformaci贸n de datos

```python
#Busca la primera columna cuyo nombre contenga alguno de los patrones dados.
def encontrar_columna(df, candidatos):

    for c in df.columns:
        nombre = c.lower()
        if any(p in nombre for p in candidatos): # 'any()' devuelve True apenas encuentra una coincidencia
            return c
    return None
```

```python
# Detectar la columna de producto
prod_col = encontrar_columna(ventas_clean, ["producto", "id_producto", "sku", "articulo", "art铆culo"])
if prod_col is None:
    raise ValueError("No se encontr贸 columna de producto.")
print(prod_col)
```

```python
# Calcular ingreso por registro = precio * cantidad

ventas_perf = (
    ventas_clean
    .assign(
        ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]
    )
)
```

```python
# Agregar m茅tricas por producto

resumen_prod = (
    ventas_perf
    .groupby(by=prod_col, dropna=False, as_index=False, observed=False)
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
    )
    .sort_values(by='ingreso_total', ascending=False)  # Ordena por ingreso_total descendente
    .assign(precio_promedio=lambda df: df['precio_promedio'].round(2))  # Redondea a 2 decimales
)


print(resumen_prod.head(10))
```

```python
# Calcular percentil 80 de ingreso_total

p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,                # q indica el percentil deseado (0.80 = 80% de los datos por debajo)
    interpolation="linear" # si el percentil no coincide exactamente con un valor real del dataset,
                           # 'linear' interpola entre los dos valores vecinos.
                           # Ejemplo: si el 80% cae entre 4000 y 5000,
                           # calcula un valor proporcional, por ejemplo 4200.
                           # Otros m茅todos posibles:
                           #  - 'lower': toma el menor de los dos valores (4000)
                           #  - 'higher': toma el mayor (5000)
                           #  - 'nearest': el m谩s cercano al percentil
                           #  - 'midpoint': el punto medio exacto (4500)
)

# En resumen:
# - quantile calcula el valor l铆mite de un percentil.
# - q define qu茅 percentil queremos.
# - interpolation define c贸mo se calcula cuando el valor no est谩 exactamente en los datos.
# El resultado (p80_ingreso) es el ingreso total que marca el l铆mite superior del 80% de los productos.


# 5锔 Filtrar los productos "de alto rendimiento" y ordenarlos
# -------------------------------------------------------------------
# Contexto: `resumen_prod` es un DataFrame con m茅tricas por producto,
# y `p80_ingreso` es el percentil 80 de la columna "ingreso_total".
# Objetivo: quedarnos con los productos cuyo ingreso_total est谩 en el 20% superior
# (ingreso_total >= p80_ingreso) y luego ordenarlos de mayor a menor por ingreso y unidades.
# en una sola fila
#ventas_top = resumen_prod.query("ingreso_total >= @p80_ingreso",engine="python").sort_values(by=["ingreso_total", "unidades"], ascending=[False, False], na_position="last", ignore_index=True
ventas_top = (
    resumen_prod
    # ---------------------------------------------------------------
    # .query(expr, *, inplace=False, engine='python'|'numexpr')
    #   - Aplica un filtro usando una expresi贸n estilo SQL-simple.
    #   - `expr` es un string que se eval煤a sobre los nombres de las columnas.
    #   - Para usar variables de Python (no columnas), se antepone '@' (ej.: @p80_ingreso).
    #   - NaN en comparaciones (>, >=, ==, etc.) se comportan como False  esas filas no pasan el filtro.
    #   - engine='python': interpreta la expresi贸n con Python puro (compatible siempre).
    #   - engine='numexpr': si est谩 instalado, acelera operaciones num茅ricas vectorizadas.
    #   - inplace: False (por defecto) devuelve un DF nuevo; True modifica el DF original (menos recomendado en cadenas).
    .query(
        "ingreso_total >= @p80_ingreso",  # expr: filtra filas donde ingreso_total es al menos el umbral del P80
        engine="python"                   # motor de evaluaci贸n (usar 'numexpr' si lo ten茅s y quer茅s performance)
        # Notas de sintaxis de `expr`:
        #    Operadores l贸gicos: and / or / not   (tambi茅n valen &, |, ~ con par茅ntesis).
        #    Strings deben ir entre comillas: canal == 'Online'
        #    Columnas con espacios o caracteres raros: usar `backticks`, ej.: `nombre producto` == 'X'
        #    Ejemplos:
        #       "ingreso_total >= @p80_ingreso and unidades >= 10"
        #       "`nombre producto`.str.contains('Promo')"
        #       "precio_promedio.between(1000, 3000, inclusive='both')"
    )
    # ---------------------------------------------------------------
    # .sort_values(by, axis=0, ascending=True|[...], inplace=False,
    #              kind='quicksort'|'mergesort'|'heapsort'|'stable',
    #              na_position='last'|'first', ignore_index=False, key=None)
    #   - Ordena por una o varias columnas.
    #   - `by`: str o lista de str con las columnas a ordenar.
    #   - `ascending`: bool o lista de bool (una por cada columna en `by`).
    #   - `na_position`: d贸nde ubicar NaN ('last' o 'first').
    #   - `ignore_index`: si True, reasigna el 铆ndice 0..n-1 en el resultado.
    #   - `kind`: algoritmo de ordenamiento (mergesort es estable si necesit谩s preservar empates).
    #   - `key`: funci贸n que transforma los valores antes de ordenar (p. ej., key=lambda s: s.str.normalize(...)).
    .sort_values(
        by=["ingreso_total", "unidades"],  # primero ordena por ingreso_total, luego desempata por unidades
        ascending=[False, False],          # ambos en orden descendente (mayor  menor)
        na_position="last",                # coloca NaN al final (煤til si alguna m茅trica qued贸 en NaN)
        ignore_index=True                  # reindexa el resultado secuencialmente (0..n-1)
        # Variantes 煤tiles:
        #    ascending=True                 # orden ascendente
        #    ascending=[False, True]        # primero desc, luego asc para el segundo criterio
        #    kind='mergesort'               # orden estable (respeta el orden de aparici贸n en empates)
        #    key=lambda s: s.str.lower()    # ordenar texto sin distinci贸n de may煤sculas/min煤sculas
    )
)

# Resultado:
# `ventas_top` contiene solo los productos cuyo ingreso_total >= p80_ingreso,
# ordenados de mayor a menor por ingreso_total y, ante empates, por unidades.

print(f"Columna de producto detectada: {prod_col}\n")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}\n")
print("Productos de alto rendimiento (top 20% por ingreso):")
display(ventas_top.head(20))
```

### 6锔 Agregaci贸n

```python
# AGREGACIN: resumen por categor铆a

def encontrar_columna(df, candidatos):

    # Recorremos todos los nombres de columnas del DataFrame (df.columns es un Index con esos nombres).
    for c in df.columns:

        # any(...) devuelve True si AL MENOS UNO de los elementos del generador interno es True.
        # Generador: (p in c.lower() for p in candidatos)
        #   - c.lower(): pasamos el nombre de la columna a min煤sculas para comparar sin importar may煤sculas/min煤sculas.
        #   - p in c.lower(): chequea si el patr贸n 'p' aparece como SUBCADENA dentro del nombre de la columna.
        #   - Se eval煤a para cada 'p' en 'candidatos'.
        if any(p in c.lower() for p in candidatos):
           # Si encontramos una coincidencia, devolvemos inmediatamente el nombre ORIGINAL de la columna.
            return c
    # Si terminamos el bucle sin encontrar coincidencias, devolvemos None para indicar "no encontrada".
    return None

# Nota did谩ctica:
# - Esto hace coincidencia por SUBCADENA (ej.: "prod" matchea "producto_sku").
# - Si quer茅s coincidencia EXACTA, us谩 igualdad (c.lower() == p) o expresiones regulares.
# - Si tus datos pueden tener acentos/espacios raros, pod茅s normalizar:
#     import unicodedata, re
#     def norm(s): return re.sub(r'\s+', ' ', ''.join(ch for ch in unicodedata.normalize('NFKD', s)
#                                    if not unicodedata.combining(ch))).strip().lower()


# 1) Detectar columna de categor铆a (acepta variantes)
cat_col = encontrar_columna(ventas_clean, ["categoria", "categor铆a", "categoria_producto", "rubro"])
if cat_col is None:
    raise ValueError("No se encontr贸 columna de categor铆a (por ej. 'categoria' o 'rubro').")

# 2) Asegurar columna 'ingreso' (si no existe, crearla)
if "ingreso" not in ventas_clean.columns:
    ventas_cat = ventas_clean.assign(ingreso = ventas_clean["precio"] * ventas_clean["cantidad"])
else:
    ventas_cat = ventas_clean.copy()

# 3) Agregaci贸n por categor铆a con groupby + agg
resumen_cat = (
    ventas_cat
    .groupby(
        by=cat_col,      # Puede ser string o lista de strings si quisi茅ramos agrupar por varias columnas.
        dropna=False,    # Mantener grupo NaN (si hay filas sin categor铆a).
        as_index=False   # Dejar la categor铆a como columna normal (y no como 铆ndice).
        # observed: si cat_col es 'category' y queremos mostrar solo categor铆as presentes  True.
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),   # Suma total de ingresos por categor铆a.
        unidades=('cantidad', 'sum'),       # Unidades totales vendidas en la categor铆a.
        ventas=('ingreso', 'size'),         # Cantidad de registros/filas (ventas) en la categor铆a.
        precio_promedio=('precio', 'mean')  # Precio promedio observado en la categor铆a.
        # Otras funciones 煤tiles: 'median','max','min','std','var','nunique'...
    )
    .sort_values(
        by='ingreso_total', # Ordenar por ingreso total
        ascending=False,    # Descendente: mayores arriba
        na_position='last', # NaN al final
        ignore_index=True   # Reindexar desde 0
    )
)

# 4) Ticket promedio por venta = ingreso_total / ventas
resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
    # assign: crea/reescribe columnas. Alternativa: resumen_cat['ticket_promedio_por_venta'] = ...
)

print("Columna de categor铆a detectada:", cat_col)
print("Resumen por categor铆a (ordenado por ingreso_total):")
display(resumen_cat.head(20))
```
